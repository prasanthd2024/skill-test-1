Write a program to work with time series data using NumPy and pandas

Problem Statement: A time series dataset representing daily temperatures over a year.
You will perform some basic analysis and visualization using NumPy, pandas, and
matplotlib.

```python
import numpy as np
import pandas as pd

# 1. Generate a time series dataset (daily temperatures over a year)
# Let's create data for a full year (365 days)
start_date = '2023-01-01'
end_date = '2023-12-31'
dates = pd.date_range(start=start_date, end=end_date, freq='D')

# Simulate daily temperatures with a general trend and some seasonality
# Base temperature
base_temp = 15
# Seasonal variation (e.g., sinusoidal for yearly cycle)
season_cycle = np.sin(np.linspace(0, 2 * np.pi, len(dates))) * 10
# Random fluctuations
random_fluctuations = np.random.normal(0, 2, len(dates))

temperatures = base_temp + season_cycle + random_fluctuations

# Create a pandas DataFrame
df_daily_temp = pd.DataFrame({
    'Date': dates,
    'Temperature': temperatures
})

df_daily_temp['Rolling_Avg_Temperature'] = df_daily_temp['Temperature'].rolling(window=7).mean()

print("DataFrame with 7-day rolling average temperature:")
display(df_daily_temp.head(10))
```
