```python
student_data = {
    "Alice": 85, "Bob": 42, "Charlie": 78, "David": 92, "Eve": 55,
    "Frank": 61, "Grace": 38, "Heidi": 89, "Ivan": 74, "Judy": 67,
    "Kevin": 50, "Liam": 49, "Mallory": 95, "Nina": 82, "Oscar": 70,
    "Peggy": 63, "Quentin": 58, "Riley": 45, "Sybil": 99, "Trent": 72
}

def analyze_class_data(data):
    scores = list(data.values())

    avg_score = sum(scores) / len(scores)
    high_score = max(scores)
    low_score = min(scores)

    print("--- CLASS SUMMARY ---")
    print(f"Average: {avg_score:.1f} | Highest: {high_score} | Lowest: {low_score}\n")

    print(f"{'Name':<10} | {'Score':<5} | {'Status':<6} | {'Grade':<5}")
    for name, score in data.items():

        status = "PASS" if score >= 50 else "FAIL"
        if score >= 90: grade = 'A'
        elif score >= 80: grade = 'B'
        elif score >= 70: grade = 'C'
        elif score >= 60: grade = 'D'
        else: grade = 'F'

        print(f"{name:<10} | {score:<5} | {status:<6} | {grade:<5}")

analyze_class_data(student_data)
```

```python
import numpy as np

def solve_linear_system(A_matrix, b_vector):
    """
    Solves the system Ax = b using NumPy.
    """
    try:
        # Convert inputs to numpy arrays
        A = np.array(A_matrix)
        b = np.array(b_vector)

        # Solve for x
        x = np.linalg.solve(A, b)
        return x

    except np.linalg.LinAlgError:
        return "Error: Matrix A is singular and cannot be solved."


A = [[3, 1],
     [1, 2]]

b = [9, 8]

solution = solve_linear_system(A, b)

print(f"Matrix A:\n{np.array(A)}")
print(f"Vector b: {b}")
print(f"Solution vector x: {solution}")
```
