### 1. Euclidean Distance for Two Points in 2D Space

The Euclidean distance between two points $(x_1, y_1)$ and $(x_2, y_2)$ in 2D space is given by the formula:

$d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$


```python
import math

def euclidean_distance(point1, point2):
    """
    Calculates the Euclidean distance between two points in 2D space.

    Args:
        point1 (tuple): A tuple (x1, y1) representing the first point.
        point2 (tuple): A tuple (x2, y2) representing the second point.

    Returns:
        float: The Euclidean distance between the two points.
    """
    x1, y1 = point1
    x2, y2 = point2
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distance

# Example usage:
point_a = (1, 2)
point_b = (4, 6)
dist = euclidean_distance(point_a, point_b)
print(f"The Euclidean distance between {point_a} and {point_b} is: {dist}")

point_c = (0, 0)
point_d = (3, 4)
dist2 = euclidean_distance(point_c, point_d)
print(f"The Euclidean distance between {point_c} and {point_d} is: {dist2}")
```

### 2. Dot Product for Two Matrices

The dot product (or matrix multiplication) of two matrices A and B is defined if the number of columns in A is equal to the number of rows in B.

If A is an `m x n` matrix and B is an `n x p` matrix, their product C = AB is an `m x p` matrix where each element $C_{ij}$ is calculated as:

$C_{ij} = \sum_{k=1}^{n} A_{ik} \cdot B_{kj}$

This means the element in the i-th row and j-th column of the product matrix is the sum of the products of the corresponding elements from the i-th row of the first matrix and the j-th column of the second matrix.


```python
def dot_product_matrices(matrix1, matrix2):
    """
    Calculates the dot product (matrix multiplication) of two matrices.

    Args:
        matrix1 (list of lists): The first matrix.
        matrix2 (list of lists): The second matrix.

    Returns:
        list of lists: The resulting dot product matrix, or None if matrices are incompatible.
    """
    rows1 = len(matrix1)
    cols1 = len(matrix1[0])
    rows2 = len(matrix2)
    cols2 = len(matrix2[0])

    # Check if matrices are compatible for multiplication
    if cols1 != rows2:
        print("Error: The number of columns in the first matrix must be equal to the number of rows in the second matrix.")
        return None

    # Initialize the result matrix with zeros
    result_matrix = [[0 for _ in range(cols2)] for _ in range(rows1)]

    # Perform matrix multiplication
    for i in range(rows1):  # iterate through rows of matrix1
        for j in range(cols2):  # iterate through columns of matrix2
            for k in range(cols1):  # iterate through columns of matrix1 / rows of matrix2
                result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]
    return result_matrix

# Example usage:
matrix_a = [
    [1, 2, 3],
    [4, 5, 6]
]

matrix_b = [
    [7, 8],
    [9, 1],
    [2, 3]
]

result = dot_product_matrices(matrix_a, matrix_b)

if result:
    print("\nMatrix A:")
    for row in matrix_a: print(row)
    print("\nMatrix B:")
    for row in matrix_b: print(row)
    print("\nDot product (A * B):")
    for row in result: print(row)

# Example of incompatible matrices:
matrix_c = [
    [1, 2],
    [3, 4]
]

matrix_d = [
    [5, 6, 7],
    [8, 9, 0],
    [1, 2, 3]
]

dot_product_matrices(matrix_c, matrix_d) # This will print an error message
```
