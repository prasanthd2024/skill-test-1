Write a program to simulate and analyze daily stock prices over one year and generate
the stock prices, calculate some basic metrics, and visualize the data with additional
features like moving averages and visualize the same.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Simulate Daily Stock Prices over one year
start_date = '2023-01-01'
end_date = '2023-12-31'
dates = pd.date_range(start=start_date, end=end_date, freq='D')
num_days = len(dates)

# Simulation parameters
initial_price = 100
drift = 0.0002  # Average daily return
volatility = 0.01  # Standard deviation of daily returns

# Generate daily returns
daily_returns = np.random.normal(drift, volatility, num_days)

# Calculate stock prices
price_series = np.zeros(num_days)
price_series[0] = initial_price

for i in range(1, num_days):
    price_series[i] = price_series[i-1] * (1 + daily_returns[i])

# Create a DataFrame
df_stock = pd.DataFrame({
    'Date': dates,
    'Price': price_series
})

print("Simulated Daily Stock Prices (first 5 rows):")
display(df_stock.head())
```

Now that we have the simulated stock prices, let's calculate some basic metrics such as daily returns, cumulative returns, and overall volatility.

```python
# 2. Calculate Basic Metrics

# Daily Returns
df_stock['Daily_Return'] = df_stock['Price'].pct_change()

# Cumulative Returns
df_stock['Cumulative_Return'] = (1 + df_stock['Daily_Return']).cumprod() - 1

# Volatility (standard deviation of daily returns)
overall_volatility = df_stock['Daily_Return'].std()

print(f"Overall Daily Volatility: {overall_volatility:.4f}")
print("Stock Data with Daily and Cumulative Returns (first 5 rows):")
display(df_stock.head())
```

Next, we'll calculate two common moving averages: a short-term (20-day) and a long-term (50-day) moving average. These are often used to identify trends and potential buy/sell signals.

```python
# 3. Calculate Moving Averages

# 20-day Simple Moving Average (SMA)
df_stock['SMA_20'] = df_stock['Price'].rolling(window=20).mean()

# 50-day Simple Moving Average (SMA)
df_stock['SMA_50'] = df_stock['Price'].rolling(window=50).mean()

print("Stock Data with Moving Averages (first 55 rows to show MA values):")
display(df_stock.head(55))
```
