Write a Python script using pandas for Student Performance Analysis:
Problem Statement:
You are given a CSV file named 'student_scores.csv' containing scores of students in
various subjects.

Student ID  Name  Maths  Physics  Chemistry

101 Alice Brown 85 88 87

102 Ben Carter 78 75 80

103 Clara Davis 92 94 90

104 David Evans 65 68 66

105 Eva Foster 74 78 76

106 Frank Green 60 55 58

107 Grace Hall 88 85 86

108 Henry Ives 70 72 74

109 Isla Jones 95 96 94

110 Jack King 82 80 78


Perform the following tasks:

a. Display the summary statistics for each subject,

b. Calculate the average score for each student,

c. Identify students who scored below 60 in more than two subjects.

```python
import pandas as pd
import io

csv_data = """Student ID,Name,Maths,Physics,Chemistry
101,Alice Brown,85,88,87
102,Ben Carter,78,75,80
103,Clara Davis,92,94,90
104,David Evans,65,68,66
105,Eva Foster,74,78,76
106,Frank Green,60,55,58
107,Grace Hall,88,85,86
108,Henry Ives,70,72,74
109,Isla Jones,95,96,94
110,Jack King,82,80,78
"""

# Save the data to a CSV file
with open('student_scores.csv', 'w') as f:
    f.write(csv_data)

print("student_scores.csv created successfully.")

# Load the CSV into a DataFrame
df_students = pd.read_csv('student_scores.csv')

display(df_students.head())
```

### Task a: Display the summary statistics for each subject.

```python
# Select only the subject columns for summary statistics
subject_scores = df_students[['Maths', 'Physics', 'Chemistry']]
display(subject_scores.describe())
```

### Task b: Calculate the average score for each student.

```python
# Calculate the average score across subjects for each student
df_students['Average Score'] = df_students[['Maths', 'Physics', 'Chemistry']].mean(axis=1)
display(df_students[['Student ID', 'Name', 'Average Score']])
```

### Task c: Identify students who scored below 60 in more than two subjects.

```python
# Define the threshold for low scores
LOW_SCORE_THRESHOLD = 60

# Count subjects where score is below the threshold for each student
df_students['Subjects Below 60'] = (df_students[['Maths', 'Physics', 'Chemistry']] < LOW_SCORE_THRESHOLD).sum(axis=1)

# Identify students who scored below 60 in more than two subjects
students_below_threshold = df_students[df_students['Subjects Below 60'] > 2]

# Display the identified students
if not students_below_threshold.empty:
    display(students_below_threshold[['Student ID', 'Name', 'Subjects Below 60', 'Maths', 'Physics', 'Chemistry']])
else:
    print("No students scored below 60 in more than two subjects.")
```
